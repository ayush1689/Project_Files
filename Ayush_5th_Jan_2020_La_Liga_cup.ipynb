### Write a function with name “Goal_diff_count” which should return all the teams with their
Goal Differences. Using the same function, find the team which has maximum and
minimum goal difference. (5 points)

```python
def goal_diff_count (f,a):
    goal_diff_count = f-a
    return goal_diff_count
```

```python
goal_diff = goal_diff_count(new_laliga['GoalsFor'],new_laliga['GoalsAgainst'])
new_laliga['goal_diff'] = goal_diff
new_laliga.head()
df1 = new_laliga[['Team','goal_diff']]
df1.head()
```

### Create a new column with the name “Winning Percent” and append it to the data set (5
points)
Percentage of Winning = (GamesWon / GamesPlayed)*100
If there are any numerical error, replace it with 0%
Print the top 5 teams which have the highest Winning percentage

```python
for ind in new_laliga.index:
    try:
        new_laliga.loc[new_laliga.index[ind],'per_winnning'] = (int(new_laliga.loc[new_laliga.index[ind],'GamesWon'])/int(new_laliga.loc[new_laliga.index[ind],'GamesPlayed']))*100
    except ZeroDivisionError:
        new_laliga.loc[new_laliga.index[ind],'per_winnning'] = '0'

new_laliga.tail()
```

```python
new_laliga.sort_values(by=['Points'],ascending=False)
df2 = new_laliga[['Team','per_winnning']]
df2.head(5)
```
